1. State the Hypotheses
Null Hypothesis (𝐻0)
There is no association between device type and customer satisfaction level. The type of device purchased and the level of satisfaction are independent.
Alternative Hypothesis (𝐻A): There is an association between device type and customer satisfaction level. The type of device purchased and the level of satisfaction are not independent.

Compute the Chi-Square Statistic
Calculate the Expected Counts:

1.The expected count for each cell in the table is calculated using:
Expected Count= Grand Total (Row Total×Column Total) 
2.Calculate the Chi-Square Statistic:

Use the formula
χ** 2 =∑ Expected Count (Observed Count−Expected Count)** 2 
3. Determine the Critical Value
Degrees of Freedom (df):

Calculate the degrees of freedom as:
df=(Number of Rows−1)×(Number of Columns−1)
df=(Number of Rows−1)×(Number of Columns−1)
For this data, we have 5 satisfaction levels and 2 device types, so 

df=(5−1)×(2−1)=4.
Critical Value:

Find the critical value for χ2  at a significance level (α) of 0.05 and 4 degrees of freedom using Chi-Square distribution tables or statistical software.

4. Make a Decision
Compare the Chi-Square Statistic with the Critical Value:
If the Chi-Square statistic is greater than the critical value, reject the null hypothesis.
 


```python
import numpy as np
import scipy.stats as stats

# Data
observed = np.array([
    [50, 70],
    [80, 100],
    [60, 90],
    [30, 50],
    [20, 50]
])

# Total counts
row_totals = np.sum(observed, axis=1)
column_totals = np.sum(observed, axis=0)
grand_total = np.sum(observed)

# Calculate expected counts
expected = np.outer(row_totals, column_totals) / grand_total

# Compute the Chi-Square statistic
chi2_statistic = np.sum((observed - expected) ** 2 / expected)

# Degrees of freedom
df = (len(row_totals) - 1) * (len(column_totals) - 1)

# Critical value for alpha = 0.05
alpha = 0.05
critical_value = stats.chi2.ppf(1 - alpha, df)

# Print results
print(f"Chi-Square Statistic: {chi2_statistic:.2f}")
print(f"Critical Value: {critical_value:.2f}")

# Decision
if chi2_statistic > critical_value:
    print("Reject the null hypothesis. There is a significant association between device type and customer satisfaction level.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between device type and customer satisfaction level.")
```

```python
# Detailed Report of Chi-Square Test Analysis
def chi_square_test():
    # Data Provided
    observed = np.array([
        [50, 70],   # Very Satisfied
        [80, 100],  # Satisfied
        [60, 90],   # Neutral
        [30, 50],   # Unsatisfied
        [20, 50]    # Very Unsatisfied
    ])

    # Step 1: State the Hypotheses
    print("Step 1: State the Hypotheses")
    print("Null Hypothesis (H0): There is no association between device type and customer satisfaction level.")
    print("Alternative Hypothesis (HA): There is an association between device type and customer satisfaction level.\n")

    # Step 2: Compute the Chi-Square Statistic
    print("Step 2: Compute the Chi-Square Statistic")

    # Total counts
    row_totals = np.sum(observed, axis=1)
    column_totals = np.sum(observed, axis=0)
    grand_total = np.sum(observed)

    # Calculate expected counts
    expected = np.outer(row_totals, column_totals) / grand_total

    # Compute the Chi-Square statistic
    chi2_statistic = np.sum((observed - expected) ** 2 / expected)

    print(f"Observed Frequencies:\n{observed}\n")
    print(f"Expected Frequencies:\n{expected}\n")
    print(f"Chi-Square Statistic: {chi2_statistic:.2f}\n")

    # Step 3: Determine the Critical Value
    print("Step 3: Determine the Critical Value")

    # Degrees of freedom
    df = (len(row_totals) - 1) * (len(column_totals) - 1)

    # Critical value for alpha = 0.05
    alpha = 0.05
    critical_value = stats.chi2.ppf(1 - alpha, df)

    print(f"Degrees of Freedom: {df}")
    print(f"Significance Level (alpha): {alpha}")
    print(f"Critical Value: {critical_value:.2f}\n")

    # Step 4: Make a Decision
    print("Step 4: Make a Decision")

    if chi2_statistic > critical_value:
        print("Decision: Reject the null hypothesis.")
        print("Conclusion: There is a significant association between device type and customer satisfaction level.")
    else:
        print("Decision: Fail to reject the null hypothesis.")
        print("Conclusion: There is no significant association between device type and customer satisfaction level.")

# Run the analysis
if __name__ == "__main__":
    chi_square_test()
```
